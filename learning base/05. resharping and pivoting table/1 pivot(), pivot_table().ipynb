```python
import pandas as pd
from pandas import read_csv
```

# pivot()


```python
# read and show the dataset
df = pd.read_csv("gdp.csv", encoding="Latin1")
```

```python
# reshape the dataframe with .pivot()
df.pivot(index = 'year', columns='country', values='gdppc')
```

# pivot_table()


```python
# read and show the dataset
df1 = pd.read_csv("supermarket_sales.csv", encoding= 'Latin1')
```

```python
# make a pivot table and add an aggregate function
df1.pivot_table(index = 'Gender', aggfunc = 'sum').round(2)
```

```python
# make a pivot table, add an aggregate function and select some columns
df1.pivot_table(index = 'Gender', columns = 'City', values = 'Total', aggfunc = 'sum').round(2)
```

```python
# make a pivot tables that says how much male and female spend in each category
df1.pivot_table(index = 'Gender', columns = 'Product line', values = 'Total', aggfunc = 'sum').round(2)
```
