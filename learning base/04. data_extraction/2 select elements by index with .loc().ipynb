```python
#  read csv file
import pandas as pd
```

```python
df_test = pd.read_csv('players_20.csv')
```

```python
df_test.set_index('short_name')
```

```python
df_test = df_test.set_index('short_name')
```

```python
df_test[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df_test = df_test[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

# selecting with single value

```python
# get all data about L.Messi
df_test.loc['L. Messi']
```

```python
# get the height of L.Messi
df_test.loc['L. Messi', 'height_cm']
```

```python
# get the weight of Cristiano Ronaldo
df_test.loc['Cristiano Ronaldo', 'weight_kg']
```

```python
# get all rows inside the 'height_cm' column
df_test.loc[:, 'height_cm']
```

```python
# get all columns that correspond to the index 'L. Messi'
df_test.loc['L. Messi', :]
```

# select with a list of values

```python
# get all data about L.Messi and Cristiano Ronaldo
df_test.loc[['L. Messi', 'Cristiano Ronaldo'], :]
```

```python
# get the height of L.Messi and Cristiano Ronaldo
df_test.loc[['L. Messi', 'Cristiano Ronaldo'] , 'height_cm']
```

```python
# get the height and weight of L.Messi
df_test.loc[['L. Messi'], ['weight_kg','height_cm']]
```

```python
# get height and weight of L.Messi and Cristiano Ronaldo
df_test.loc[['L. Messi', 'Cristiano Ronaldo'] , ['weight_kg','height_cm']]
```

# selecting a range of data with slice

```python
# slice column labels
player = ['L. Messi', 'Cristiano Ronaldo']
df_test.loc[player, 'long_name':'club']
```

```python
# slice index labels
df_test.index[:10]
```

```python
df_test.loc['L. Messi':'M. Salah', :]
```

# select with conditions

```python
df_test['height_cm'] > 180
```

```python
# one condition: select player with height above 180cm
df_test.loc[df_test['height_cm'] > 180, 'long_name']
```

```python
# multiple conditions: select player with height above 180cm from Argentina
df_test.loc[(df_test['height_cm'] > 180) & (df_test['nationality'] == 'Portugal')]
```
