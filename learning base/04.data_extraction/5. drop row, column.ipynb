```python
from operator import index

import pandas as pd
from sqlalchemy import true

df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

# Drop Row

```python
# drop one row
# axis parameter
df.drop('J. Oblak', axis=0, inplace=True)
```

```python
# index parameter
df.drop(index= ['Shao Shuai'], inplace=True)
```

```python
# drop two or more rows and update data (inplace=True)
df.drop(['Cristiano Ronaldo', 'L. Messi'], axis = 0, inplace=True)
```

# Drop Columns


```python
# drop one column
# axis parameter
df.drop('age', axis=1)
```

```python
# index parameter
df.drop(columns = 'dob')
```

```python
# drop column by position/index (drop last column)
df.drop(df.columns[-1], axis = 1)
```

```python
# drop two or more columns and update data (inplace=True)
df.drop(['age', 'dob', 'nationality'], axis = 1, inplace = True)
```
