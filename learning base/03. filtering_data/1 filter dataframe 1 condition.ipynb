# filter data use np.where

```python
import numpy as np
```

```python
# create an array based on price tiers
np.where(df_laptop.Price_euros > 2000, 'Expensive', 'Cheap')
```

```python
# set it to a new column / add new column into dataframe laptop
df_laptop['price_tiers'] = np.where(df_laptop.Price_euros > 2000, 'Expensive', 'Cheap')
```

```python
df_laptop.head()
```

```python
#counting price_tiers
df_laptop.value_counts('price_tiers')
```
