```python
import pandas as pd
```

```python
import numpy as np
```

```python
df_laptop = pd.read_csv('laptop_price.csv', encoding='Latin1')
```

# create a condition column
### tạo 1 column mới bởi giá trị price , inch using np.select(condition,...)

```python
# create an array based on multiple price tiers (+2 choices)


# create "conditions" and "values" variables
conditions = [
    df_laptop['Price_euros'] > 3000,
    (df_laptop['Price_euros'] > 2000) & (df_laptop['Price_euros'] <= 3000),
    (df_laptop['Price_euros'] > 800) & (df_laptop['Price_euros'] < 2000),
    df_laptop['Price_euros'] <= 800
]

values = ['really expensive', 'expensive', 'affordable', 'cheap']
```

```python
# set it to a new column
df_laptop['price_tiers']= np.select(conditions, values, default= "other")
```

```python
df_laptop
```

```python
# count values icnside "price_tier" column
df_laptop['price_tiers'].value_counts()
```

## excercise

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small<12

# create "conditions" and "values" variables
condition = [
    df_laptop['Inches'] > 16,
    (df_laptop['Inches'] <= 16) & (df_laptop['Inches'] > 14),
    (df_laptop['Inches'] <= 14) & (df_laptop['Inches'] > 12),
    df_laptop['Inches'] < 12
             ]
```

```python
value = [
    'Too Big' , 'Big', 'Small', 'Too Small'
]
```

```python
# set it to a new column
df_laptop['size_tiers'] = np.select(condition, value, default= "other")
```
