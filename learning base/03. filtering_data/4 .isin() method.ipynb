```python
import pandas as pd
```

```python
df_laptop = pd.read_csv('laptop_price.csv', encoding='Latin1')
```

###### (introduce data in form of list [])

# isin() : single filtering

```python
# select Apple or HP Laptops
df_laptop['Company'].isin(['Hp', 'Apple'])
```

```python
# filter dataframes
df_laptop[df_laptop['Company'].isin(['HP', 'Apple'])].value_counts('Company')
```

```python
# find notebooks or ultrabooks from Apple or HP
# multiple conditions
find1 = df_laptop['Company'].isin(['HP', 'Apple'])
find2 = df_laptop['TypeName'].isin(['Ultrabook', 'Notebook'])
```

```python
# filter dataframe
df_laptop[find1 & find2].value_counts('Company')
```

```python
# find notebooks or ultrabooks from Apple or HP
# multiple conditions
find1 = df_laptop['Company'].isin(['HP', 'Apple'])
find2 = df_laptop['TypeName'].isin(['Ultrabook', 'Notebook'])
```

```python
# filter dataframe
df_laptop[find1 & find2].value_counts('Company')
```
