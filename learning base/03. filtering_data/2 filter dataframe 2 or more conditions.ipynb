# filter dataframe 2 or many conditions

```python
df_laptop = pd.read_csv('laptop_price.csv', encoding='Latin1')
```

```python
# find Apple laptops that cost more than 2000 euros
(df_laptop.Company == 'Apple') & (df_laptop.Price_euros > 2000)
```

```python
#use & and () to connect condition to each other, we also can use and and or & |
df_laptop[(df_laptop.Company == 'Apple') & (df_laptop.Price_euros > 2000)]
```

```python
# find laptops Apple or Dell laptops
df_laptop[(df_laptop.Company == 'Apple') | (df_laptop.Company == 'Dell')].value_counts('Company')
```

```python
# filer dataframe based on multiple conditions
# find laptops from Apple or Dell that cost over 2000 euros
df_laptop[((df_laptop.Company == 'Apple') | (df_laptop.Company == 'Dell')) & (df_laptop.Price_euros > 2000)].value_counts('Company')
```

##### & is and, | is or and we must have parenthesis while connecting two conditions
