```python
df = pd.read_csv("population_total.csv")
```

```python
df
```

# Making a Pivot Table


```python
# dropping null values
df.dropna(inplace=True)
```

```python
# making a pivot table
df = df.pivot(columns= "country", index= 'year', values= 'population')
```

```python
# selecting some countries
df = df[['China', 'France', 'United States', 'Vietnam', 'India']]
```

```python
# showing pivot table
df
```

# Line plot

```python
df.plot(kind='line', xlabel='year', ylabel='population', title='Population by year', figsize= (8,4))
```

## Bar plot


### Single Barplot

```python
# selecting only one year (2020)
df2005 = df[df.index.isin([2005])]
```

```python
# showing new dataframe
df2005
```

```python
# transpose dataframe from rows to columns or viceversa USE .T
df2005 = df2005.T
```

```python
# making barplot
df2005.plot(kind='bar',
            color='green',
            xlabel='Year',
            ylabel='Population',
            figsize= (8,4))
```

## Multiple Barplot

```python
# selecting a few years
df = df[df.index.isin([2005, 1995, 2000,2010, 2020])]
```

```python
# making grouped barplot
df.plot(kind='bar',color = 'green',xlabel='year', ylabel='population', title='Population by year', figsize= (8,4))
```

# Piechart

```python
# changing column name
df2005.rename(columns={2005:'2005'}, inplace=True)
```

```python
# making the plot
df2005.plot(kind='pie',y = '2005' ,title='Population in 2005 (%)')
```

# Boxplot


## Single Boxplot


```python
df['France'].plot(kind = "box", ylabel = 'population', title='Population in France' ,figsize= (8,4))
```

## Multiple Boxplot


```python
df.plot(kind="box", xlabel='Countries',ylabel = 'population', title='Population sample for 5 countries')
```

# Histogram


```python
df.plot(kind= 'hist', xlabel='Population')
```

```python
df[['France', 'Vietnam']].plot(kind = 'hist', xlabel='Population')
```

# ScatterPlot


```python


# selecting a sample of the raw data
df.plot(kind = 'scatter', x = 'year', y = 'population')
```
