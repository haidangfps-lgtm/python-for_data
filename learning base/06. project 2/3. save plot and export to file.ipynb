```python
import pandas as pd
from matplotlib_inline.backend_inline import show
```

```python
import matplotlib.pyplot as plt
```

```python
    # reading the csv file
df = pd.read_csv("population_total.csv")
```

```python
# first look at the dataset
df
```

# Making a Pivot Table


```python
# dropping null values
df.dropna(inplace=True)
```

```python
# making a pivot table
df = df.pivot(columns='country', index='year',values='population')
```

```python
df
```

```python
# selecting some countries
df = df[['Vietnam', 'France']]
```

```python
# showing pivot table
df
```

```python
# export pivot table
df.to_excel('French and VietNam.xlsx')
```

# Lineplot


```python
df
```

```python
# save plot
df.plot(kind='line', xlabel='Year', ylabel='Population', title='Population vs Year')
plt.savefig('test.png')
plt.show()
```
