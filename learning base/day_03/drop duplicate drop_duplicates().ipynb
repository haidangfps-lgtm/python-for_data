```python
import pandas as pd
df_laptop = pd.read_csv('laptop_price.csv', encoding='Latin1')
```

# drop duplicates

```python
# drop duplicates in 2 or more columns
df_laptop.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptop = df_laptop.sort_values(['Company', 'Price_euros'])
```

```python
# cheapest: keep='first'
df_laptop.drop_duplicates(['Company'], keep = 'first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last'
df_laptop.drop_duplicates(['Company'], keep = 'last')[['Company', 'Price_euros']]
```

```python
# Argument
# inplace: drop duplicates in place or to return a copy.
# ignore_index=True: the resulting axis will be labeled 0, 1, â€¦

df_laptop.drop_duplicates(['Company'], keep = 'first', inplace=True, ignore_index=True)
```
