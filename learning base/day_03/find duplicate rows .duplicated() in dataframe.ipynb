# duplicate row using keep

Task: Find the cheapest and most expensive laptop of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)


```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptop = df_laptop.sort_values(['Company', 'Price_euros'])
```

```python
# check all value categories in "Company" column
df_laptop['Company'].value_counts()
```

```python
# duplicated values in "Company" columns (default -> keep='first')
df_laptop.duplicated(['Company'], keep = 'first')
```

```python
# show dataframe with duplicated values in "Company" column
# keep='first'(cheapest laptop per company)
duplicate_first = df_laptop.duplicated(['Company'], keep = 'first')
# show dataframe with non duplicated values in "Company" column
df_laptop[~duplicate_first][['Company', 'Price_euros']].value_counts()
```

```python
# check all value categories
df_laptop[~duplicate_first].value_counts('Company')
```

```python
# keep='last
duplicate_last = df_laptop.duplicated(['Company'], keep = 'last')
```

```python
# keep=False
duplicate_false = df_laptop.duplicated(['Company'], keep = False)
```

```python
# keep='last'(most expensive laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptop[~duplicate_last].value_counts(['Company', 'Price_euros'])
```

```python
# keep=False
# show dataframe with non duplicated values in "Company" column
df_laptop[~duplicate_false].value_counts(['Company', 'Price_euros'])
```

```python
# all elements in company column are repeated at least once
df_laptop.value_counts('Company')
```
