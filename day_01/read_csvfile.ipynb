```python
# root = https://www.football-data.co.uk/mmz4281/
# season = 2526
# league = E0
```

```python
#create root
root = 'https://www.football-data.co.uk/mmz4281/'
```

```python
# creating list of leagues
leagues = ['E0', 'E2', 'E3']
```

```python
# looping through leagues, read multiple csv and append it into a list
multiple_leauges = []
for i in leagues:
    df = pd.read_csv(root + '2526' + '/' + i + '.csv')
    multiple_leauges.append(df)
```

### 2.2 multiple seasons

```python
#create one more loop inside multiple leauge to create the new multiple seasons
#range 2018 - 2026
multiple_leauges = []
for i in leagues:
    for j in range(20,26):
        df = pd.read_csv(root + str(j) + str(j+1) + '/' + i + '.csv')
        df.insert(1, 'j', j)
        multiple_leauges.append(df)
```

# 3 organize data in a dictionary (key, value)


```python
# create a dictionary with original name of league as key
dict_countries = {
    'Spanish La Liga': 'SP1',
    'Spanish Segunda Division': 'SP2',
    'German Bundesliga': 'D1',
    'English Premier League': 'E0',
    'English League 1': 'E2',
    'English League 2': 'E3',
}
```

```python
dictionary_data = {}

for i in dict_countries:
    multiple_leauges = []
    for j in range(20,26):
        df = pd.read_csv(root + str(j) + str(j+1) + '/' + dict_countries[i] + '.csv')
        df.insert(1, 'j', j)
        multiple_leauges.append(df)
    df_data = pd.concat(multiple_leauges)
    dictionary_data[i] = df_data
```
